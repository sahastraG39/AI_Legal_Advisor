# Project 1: Text to Speech using AI

#This project converts text input into speech using a pre-trained
#Text-to-Speech (TTS) model.


```python
!pip install gtts
```

```python
from gtts import gTTS
from IPython.display import Audio
```

```python
text = "Hello, this is my first AI text to speech project."
```

```python
tts = gTTS(text=text, lang='en')
tts.save("output.mp3")
```

```python
Audio("output.mp3")
```

## Workflow

1. User provides text input
2. Text is sent to a pre-trained TTS model
3. Model converts text into speech
4. Audio file is generated and played
